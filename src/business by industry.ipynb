```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv('csv/2020 employment by industry.csv', thousands = ',')
```

```python
df = df.groupby(['area_title', "industry_title"])['qtrly_estabs_count'].std()
df = df.add_suffix('_Number of units normalized').reset_index()
df = df.rename(columns = {'area_title': 'state', 'industry_title' : 'industry', 'qtrly_estabs_count':'number of industry units normalized'})
df
```

```python
df = df.pivot(index="state", columns = 'industry', values ="number of industry units normalized")
```

```python
df
```

```python
df = df.filter(['102 Service-providing_Number of units normalized','NAICS 23 Construction_Number of units normalized','1023 Financial activities_Number of units normalized', '1013 Manufacturing_Number of units normalized', 'NAICS 51 Information_Number of units normalized'])
```

```python
df.to_csv('units by industry.csv')
```
