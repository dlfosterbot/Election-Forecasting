```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('csv/us-states-covid.csv', thousands = ',')
```

```python
res = df[~(df['date'] > '2020-11-06')]
```

```python
res = res.sort_values(by=['state'])
res = res.groupby(['state']).mean()
res = res.reset_index()
res
```

```python
res.to_csv('clean-states-covid.csv', index = False, header = True)
```
